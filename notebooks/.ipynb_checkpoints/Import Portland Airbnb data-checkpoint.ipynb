```python
import pandas as pd
import re
from datetime import datetime

from bs4 import BeautifulSoup, SoupStrainer
import requests
import pickle

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "http://insideairbnb.com/get-the-data.html"

page = requests.get(url)    
data = page.text
soup = BeautifulSoup(data)

links = []
for link in soup.find_all('a'):
    links.append(link.get('href'))
```

```python
def gather_links(city,cat):

    len(links)
    links_df = pd.DataFrame(links, columns = ['link'])
    links_df2 = links_df[links_df['link'].str.contains(city)==True]
    links_df3 = links_df2[links_df2['link'].str.contains(cat)==True]
    return links_df3
```

```python
def readin_gz(dt,cat,state='or',city='portland'):
    url = 'http://data.insideairbnb.com/united-states/{}/{}/{}/data/{}.csv.gz'.format(str(state),str(city),str(dt),str(cat))

    df = pd.read_csv(url,compression='gzip',header=0, sep=',', quotechar='"',dtype= {'zipcode':object})
    return df
    
```

```python
links_sf_listings = gather_links('portland','listings.csv.gz')
links_sf_listings['dt'] = [x.split('/')[6] for x in links_sf_listings['link']]
links_sf_listings

idx = links_sf_listings['dt'].unique()
len(idx),idx
```

```python
listings = readin_gz(idx[0],'listings')    
listings['file_scraped_dt'] = idx[0]
listings.head(1)
```

```python
# broken link for Oct-3-2016

for i, dt in enumerate(idx[1:]):
    
    if dt == '2016-10-03':
        print('bad link {}'.format(dt))
        pass
    elif dt == '2015-10-01':
        print('bad link {}'.format(dt))
        pass
    else:
        
        df = readin_gz(idx[i+1],'listings')
   
        df['file_scraped_dt'] = dt
        print(dt,df['last_scraped'].value_counts())
        listings = pd.concat([listings, df], ignore_index = True,axis = 0,sort=False)
        
```

```python
ls_cols = listings.columns

listings['is_duplicated'] = listings.duplicated(ls_cols)
listings['is_duplicated'].sum()

listings_u = listings[listings['is_duplicated'] == False].drop(['is_duplicated'],axis=1)
listings_u.shape
```

```python
links_sf_calendar = gather_links('portland','calendar.csv.gz')
links_sf_calendar['dt'] = [x.split('/')[6] for x in links_sf_calendar['link']]
links_sf_calendar.head()

idx = links_sf_calendar['dt'].unique()
len(idx),idx
```

```python
# limit to observations where "available" indicates True

calendar = readin_gz(idx[0],'calendar')     
calendar['file_scraped_dt'] = idx[0]

calendar['date'] = pd.to_datetime(calendar['date'])
calendar['file_scraped_dt'] = pd.to_datetime(calendar['file_scraped_dt'])
calendar['lead_time'] = (calendar['date'] - calendar['file_scraped_dt']).dt.days
calendar = calendar[calendar['available'] == 't']

```

```python
for i, dt in enumerate(idx[1:]):
    if dt == '2016-10-03':
        print('bad link {}'.format(dt))
        pass
    elif dt == '2015-10-01':
        print('bad link {}'.format(dt))
        pass
    else:
        df = readin_gz(idx[i+1],'calendar')
        df['file_scraped_dt'] = dt

        df['date'] = pd.to_datetime(df['date'])
        df['file_scraped_dt'] = pd.to_datetime(df['file_scraped_dt'])
        df['lead_time'] = (df['date'] - df['file_scraped_dt']).dt.days

        df2 = df[df['available'] == 't']

        print(dt, df2['file_scraped_dt'].value_counts())
        calendar = pd.concat([calendar, df2], ignore_index = True,axis = 0,sort=False)
```

```python

calendar.shape 
```

```python
cal_cols = calendar.columns
cal_cols
calendar.head()
```

```python
calendar['is_duplicated'] = calendar.duplicated(cal_cols)
calendar['is_duplicated'].sum()

calendar_u = calendar[calendar['is_duplicated'] == False].drop(['is_duplicated'],axis=1)
calendar_u.shape
```
