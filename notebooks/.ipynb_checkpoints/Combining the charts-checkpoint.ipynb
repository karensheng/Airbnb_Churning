```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

import datetime
```

```python
price = pd.read_csv('../data/median_comp_price.csv')
price.drop('date.1', axis = 1, inplace=True)
price['date'] = pd.to_datetime(price['date'])
price = price[69:]
price.head()
```

```python
# Inferences from CausalImpact

ci_nov2017 = pd.read_csv('../data/causalimpact_infer_nov52017_median.csv')
ci_nov2017 = ci_nov2017.rename(columns={"Unnamed: 0": "date"})
ci_nov2017['date'] = pd.to_datetime(ci_nov2017['date'])
ci_nov2017 = ci_nov2017[69:]
```

```python
# predicted values from forecasting

forecast_nov2017 = pd.read_csv("../data/forecast_sf_median_114.csv")
forecast_nov2017['ds'] = pd.to_datetime(forecast_nov2017['ds'])
forecast_nov2017 = forecast_nov2017[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
forecast_nov2017 = forecast_nov2017[69:]
```

```python
forecast_nov2017_portland = pd.read_csv("../data/forecast_portland_median_114.csv")
forecast_nov2017_portland['ds'] = pd.to_datetime(forecast_nov2017_portland['ds'])
forecast_nov2017_portland = forecast_nov2017_portland[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]

forecast_nov2017_portland = forecast_nov2017_portland[69:]



```

```python
# Show observed values only from both cities

fig, ax = plt.subplots(1,figsize = (15,10))

xlinspace = pd.date_range('2017-01-01', periods= price.shape[0], freq='W')

sns.lineplot(x= 'date', y='SF_median_price', color='k', data = price, label = "Observed SF Price")

sns.lineplot(x= 'date', y='Portland_median_price', data = price, color='c',\
             label = "Observed Portland Price")

plt.axvline(x= datetime.date(2017,11,5), color = 'purple', linestyle='--',label='Cutoff for Intervention: Nov-5-2017')


ax.set_ylim(0,200)
plt.xlabel("Date \n\n\nThe data from Sept-6-2015 to Nov-5-2017 were used as the pre-period.", fontsize=20)
plt.ylabel("Price $", fontsize=20)
plt.title("Weekly Median Price of Private Room Listings: San Francisco vs Portland \n \n Jan 2017 - June 2019", fontsize = 25)
plt.legend(loc=1, prop={'size': 13})

xfmt = mdates.DateFormatter('%m-%d-%y')


plt.tight_layout()

plt.savefig('../images/observed_median_price_only.png')
```

```python
# Show counterfactual price from BSTS


fig, ax = plt.subplots(1,figsize = (15,10))

xlinspace = pd.date_range('2017-01-01', periods= price.shape[0], freq='W')

sns.lineplot(x= 'date', y='SF_median_price', color='k', data = price, label = "Observed SF Price")

sns.lineplot(x= 'date', y='post_preds', data = ci_nov2017, color = 'b', \
             label = "Predicted SF Price from BSTS")

ax.lines[1].set_linestyle("--")


plt.fill_between(ci_nov2017['date'],ci_nov2017['post_preds_lower'], ci_nov2017['post_preds_upper'], 
                facecolor = 'blue',
                interpolate = 'True',
                alpha = 0.35)


sns.lineplot(x= 'date', y='Portland_median_price', data = price, color='c',\
             label = "Observed Portland Price")


plt.axvline(x= datetime.date(2017,11,5), color = 'purple', linestyle='--',label='Cutoff for Intervention: Nov-5-2017')


ax.set_ylim(0,200)
plt.xlabel("Date \n\n\nThe data from Sept-6-2015 to Nov-5-2017 were used for the pre-period.", fontsize=20)
plt.ylabel("Price $", fontsize=20)
plt.title("Weekly Median Price of Private Room Listings: San Francisco vs Portland \n \n Jan 2017 - June 2019", fontsize = 25)
plt.legend(loc=1, prop={'size': 13})

xfmt = mdates.DateFormatter('%m-%d-%y')

plt.tight_layout()

plt.savefig('../images/CausalImpact_Only_Nov2017.png')
```

```python
# Show forecasting and CIs

fig, ax = plt.subplots(1,figsize = (15,10))

xlinspace = pd.date_range('2017-01-01', periods= price.shape[0], freq='W')

sns.lineplot(x= 'date', y='SF_median_price', color='k', data = price, label = "Observed SF Price")


sns.lineplot(x = 'ds', y = 'yhat', data = forecast_nov2017, color = 'red', \
             label = 'Predicted SF Price from forecasting')

ax.lines[1].set_linestyle("--")

plt.fill_between(forecast_nov2017['ds'],forecast_nov2017['yhat_lower'],
                 forecast_nov2017['yhat_upper'], 
                facecolor = 'red',
                interpolate = 'True',
                alpha = 0.25)


sns.lineplot(x= 'date', y='Portland_median_price', data = price, color='c',\
             label = "Observed Portland Price")
sns.lineplot(x = 'ds', y='yhat', data = forecast_nov2017_portland, \
             color = 'red', \
             label = 'Predicted Portland Price from forecasting')
ax.lines[3].set_linestyle('--')
plt.fill_between(forecast_nov2017_portland['ds'],forecast_nov2017_portland['yhat_lower'],
                 forecast_nov2017_portland['yhat_upper'], 
                facecolor = 'red',
                interpolate = 'True',
                alpha = 0.25)

    
plt.axvline(x= datetime.date(2017,11,5), color = 'purple', linestyle='--',label='Cutoff for Intervention: Nov-5-2017')


ax.set_ylim(0,200)
plt.xlabel("Date \n\n\nThe data from Sept-6-2015 to Nov-5-2017 were used for the pre-period.", fontsize=20)
plt.ylabel("Price $", fontsize=20)
plt.title("Weekly Median Price of Private Room Listings: San Francisco vs Portland \n \n Jan 2017 - June 2019", fontsize = 25)
plt.legend(loc=1, prop={'size': 13})

xfmt = mdates.DateFormatter('%m-%d-%y')

plt.tight_layout()

plt.savefig('../images/Forecast_Only_Nov2017.png')
```

```python
# Show everything!


fig, ax = plt.subplots(1,figsize = (15,10))

xlinspace = pd.date_range('2017-01-01', periods= price.shape[0], freq='W')

sns.lineplot(x= 'date', y='SF_median_price', color='k', data = price, label = "Observed SF Price")

sns.lineplot(x= 'date', y='post_preds', data = ci_nov2017, color = 'b', \
             label = "Predicted SF Price from BSTS")

ax.lines[1].set_linestyle("--")


plt.fill_between(ci_nov2017['date'],ci_nov2017['post_preds_lower'], ci_nov2017['post_preds_upper'], 
                facecolor = 'blue',
                interpolate = 'True',
                alpha = 0.35)

sns.lineplot(x = 'ds', y = 'yhat', data = forecast_nov2017, color = 'red', \
             label = 'Predicted SF Price from forecasting')

ax.lines[2].set_linestyle("--")

plt.fill_between(forecast_nov2017['ds'],forecast_nov2017['yhat_lower'],
                 forecast_nov2017['yhat_upper'], 
                facecolor = 'red',
                interpolate = 'True',
                alpha = 0.25)



sns.lineplot(x= 'date', y='Portland_median_price', data = price, color='c',\
             label = "Observed Portland Price")
sns.lineplot(x = 'ds', y='yhat', data = forecast_nov2017_portland, \
             color = 'red', \
             label = 'Predicted Portland Price from forecasting')
ax.lines[4].set_linestyle('--')
plt.fill_between(forecast_nov2017_portland['ds'],forecast_nov2017_portland['yhat_lower'],
                 forecast_nov2017_portland['yhat_upper'], 
                facecolor = 'red',
                interpolate = 'True',
                alpha = 0.25)

    
plt.axvline(x= datetime.date(2017,11,5), color = 'purple', linestyle='--',label='Cutoff for Intervention: Nov-5-2017')


ax.set_ylim(0,200)
plt.xlabel("Date \n\n\nThe data from Sept-6-2015 to Nov-5-2017 were used for the pre-period.", fontsize=20)
plt.ylabel("Price $", fontsize=20)
plt.title("Weekly Median Price of Private Room Listings: San Francisco vs Portland \n \n Jan 2017 - June 2019", fontsize = 25)
plt.legend(loc=1, prop={'size': 13})

xfmt = mdates.DateFormatter('%m-%d-%y')

plt.tight_layout()
plt.savefig('../images/CI_Forecast_Combined_Nov2017.png')
```

```python
fig, ax = plt.subplots(1,figsize = (15,8))

xlinspace = pd.date_range('2015-09-06', periods= price.shape[0], freq='W')

sns.lineplot(x= 'date', y='point_effects', data = ci_nov2017, color = 'blue', \
             label = "Estimated Pointwise Effects")

ax.lines[0].set_linestyle("--")

plt.fill_between(ci_nov2017['date'],ci_nov2017['point_effects_lower'],\
                 ci_nov2017['point_effects_upper'], \
                facecolor = 'blue',
                interpolate = 'True',
                alpha = 0.35)
plt.axvline(x= datetime.date(2017,11,5), color = 'purple', \
            linestyle='--',label='Cutoff for Intervention: Nov-5-2017')
plt.xlabel('Date', fontsize=20)
plt.ylabel('Point Effects $', fontsize=20)
plt.title('Estimated Pointwise Effects of Market Intervention',fontsize=25)
plt.legend(loc=1, prop={'size': 13})

plt.tight_layout()
plt.savefig('../images/point effects_nov2017.png')
plt.show
```
