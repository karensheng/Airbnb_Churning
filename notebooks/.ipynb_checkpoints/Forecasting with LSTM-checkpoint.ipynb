```python
import tensorflow as tf
keras = tf.keras
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

%matplotlib inline

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import *

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from keras.callbacks import EarlyStopping
```

```python
price = pd.read_csv('../data/median_comp_price.csv')
price = price.fillna(method= 'ffill')
# price = price[1:]
sf_price = price[['date', 'SF_median_price']]
portland_price = price[['date', 'Portland_median_price']]
sf_price.columns = ['ds', 'y']
portland_price.columns = ['ds', 'y']
```

```python
train, test = sf_price[1:132], sf_price[132:]
```

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return np.asarray(X), np.asarray(y)
 
# define input sequence
raw_seq = sf_price['y'].to_list()
# choose a number of time steps
n_steps = 20
# split into samples
X_train, y_train = split_sequence(train['y'].to_list(), n_steps)
X_test, y_test = split_sequence(test['y'].to_list(), n_steps)

# # summarize the data
# for i in range(len(X)):
# 	print(X[i], y[i])
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
# model = Sequential()
# model.add(LSTM(100, input_shape=(X_train.shape[1], X_train.shape[2])))
# model.add(Dropout(0.2))
# model.add(Dense(1))
# model.compile(loss='mean_squared_error', optimizer='adam')

# history = model.fit(X_train, y_train, epochs=100, batch_size=70, validation_data=(X_test, y_test), 
#                     callbacks=[EarlyStopping(monitor='val_loss', patience=10)], verbose=1, shuffle=False)


model = keras.Sequential()
model.add(keras.layers.LSTM(60, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(Dropout(0.1))
model.add(keras.layers.LSTM(120, return_sequences=False))
model.add(Dropout(0.1))
model.add(keras.layers.Dense(1, activation='linear'))
model.compile(optimizer='adam',
              loss='mse')
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=5000)
```

Let's see what predictions we get.

```python
y_pred = model.predict(X_test)
```

```python
# make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```
