```python
import statsmodels.api as sm

reviews_tab = reviews.resample('W').size().reset_index(name='Counts') 
reviews_tab = reviews_tab.set_index(pd.DatetimeIndex(reviews_tab['date']))

reviews_tab_post2015 = reviews_tab['2015-05-01':]

decomposition = sm.tsa.seasonal_decompose(reviews_tab_post2015.dropna()['Counts'].values, model='additive', freq= 52)  
booking_decomp = reviews_tab_post2015.dropna().copy()
booking_decomp['trend'] = decomposition.trend
booking_decomp['seasonal'] = decomposition.seasonal
booking_decomp['resid'] = decomposition.resid

fig, axs = plt.subplots(4,1, figsize=(10,10), sharex=True)
columns = ['Counts','trend','seasonal','resid']

for i, ax in enumerate(axs):
    ax.set_title('Decompostion of Weekly Bookings on Airbnb: 2009 - 2019')
    ax.plot(booking_decomp.index, booking_decomp[columns[i]].values, lw=.5)
    ax.set_ylabel(columns[i].capitalize(), size=10)
```

```python
grouper = reviews_sub.groupby([pd.Grouper(freq='1M'), 'std_nb'])
demand_loc = grouper.size().reset_index(name = 'counts')

demand_loc.head()
```

```python
lst = ['Mission', 'South of Market', 'Financial District', 'Western Addition', 'Downtown/Civic Center', 'Pacific Heights']


for_chart = demand_loc[demand_loc['std_nb'].isin(lst)]

fig, ax = plt.subplots(1,1,figsize = (15,10))

sns.lineplot(x= 'date', y='counts', hue = 'std_nb', data = for_chart)
# Dreamforce dates 2015 - 2018

plt.axvline(x= datetime.date(2015,9,18), color = 'red', linestyle='--')
plt.axvline(x= datetime.date(2016,10,7), color = 'red', linestyle='--')
plt.axvline(x= datetime.date(2017,11,9), color = 'red', linestyle='--')
plt.axvline(x= datetime.date(2018,9,28), color = 'red', linestyle='--')

# Outside Lands Festival
plt.axvline(x= datetime.date(2018,8,12), color = 'c', linestyle='--')

# Superbowl in 2016
plt.axvline(x= datetime.date(2016,2,7), color = 'black', linestyle='--')

plt.title("Monthly Demand of Airbnb Homes in Selected SF Neighborhoods: May 2015 - June 2019")
plt.xlabel("Date")
plt.ylabel("Number of Reviews");
```
